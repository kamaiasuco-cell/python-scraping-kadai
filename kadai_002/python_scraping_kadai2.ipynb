```python
import requests
import pprint
from getpass import getpass


api_key = getpass("Google Maps APIキーを入力してください: ")


station_name = input("駅名を入力してください（例：新宿駅）: ")


geocode_url = "https://maps.googleapis.com/maps/api/geocode/json"

geocode_response = requests.get(
    geocode_url,
    params={
        "address": station_name,
        "key": api_key
    }
)

geocode_data = geocode_response.json()


if geocode_data["status"] != "OK":
    print("駅の位置情報が取得できませんでした。")
    exit()

location = geocode_data["results"][0]["geometry"]["location"]
latitude = location["lat"]
longitude = location["lng"]

print(f"\n取得した緯度: {latitude}")
print(f"取得した経度: {longitude}")


places_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

places_response = requests.get(
    places_url,
    params={
        "location": f"{latitude},{longitude}",  # ← Nearby Searchではlocation指定
        "radius": 500,
        "type": "restaurant",
        "language": "ja",
        "key": api_key
    }
)

places_data = places_response.json()



if places_data["status"] != "OK":
    print("レストラン情報が取得できませんでした。")
    exit()


print("\n【半径500m以内のレストラン】\n")

for place in places_data["results"]:
    name = place.get("name", "不明")
    rating = place.get("rating", "評価なし")
    address = place.get("vicinity", "住所不明")

    print(f"店名: {name}")
    print(f"評価: {rating}")
    print(f"住所: {address}")
    print("-" * 40)
```
